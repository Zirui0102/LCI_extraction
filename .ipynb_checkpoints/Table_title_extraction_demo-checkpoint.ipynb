```python
#extract paper titles from a list of file paths
def extract_titles(file_paths):
    
    titles = []
    for path in file_paths:
        # Extract the file name without the extension
        match = re.search(r'([^/\\]+)\.pdf$', path)
        if match:
            title = match.group(1)
            # Remove the specified suffix if present
            titles.append(title)
    
    return pd.DataFrame(titles, columns=['Paper_title'])
```

```python
#define a function to extract table titles
def table_titles(pdf_path):
    def extract_table_titles(pdf_path):
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            num_pages = len(reader.pages)
            table_titles = []
    
            # Regular expression to find table titles
            table_title_regex = re.compile(r'[Tt][Aa][Bb][Ll][Ee]+\.?+\s*[A-Z]?\d+\.?\s+[A-Z]+.*')
    
            for i in range(num_pages):
                page = reader.pages[i]
                text = page.extract_text()
                if text:
                    found_titles = table_title_regex.findall(text)
                    table_titles.extend(found_titles)
            return table_titles
            
    table_titles = extract_table_titles(pdf_path)
    single_cell_content = ' | '.join(table_titles)
    df = pd.DataFrame([[single_cell_content]], columns=['Table_title'])
    df['Table_title'] = df['Table_title'].str.replace(r'\n', ' ', regex=True)
                
    return df
```

```python
def get_embedding(text, model="text-embedding-ada-002"):
    text = text.replace("\n", " ")
    return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
topics = [
    {'label': 'Inventory data', 'description': 'inventory data or input and output data'},
    {'label': 'Other data', 'description': 'other types of data, such as LCA results and impact factors'}
]
```

```python
label_descriptions = [
    "Inventory data or input-output data",
    "LCA results, impact factors, parameters"
]

# Assuming you have the `get_embedding` function as defined previously
label_embeddings = [get_embedding(description) for description in label_descriptions]
```

```python
def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)
```

```python
def find_most_similar_label(row_embedding, label_embeddings):
    similarities = [cosine_similarity(row_embedding, label_embedding) for label_embedding in label_embeddings]
    # Assuming labels are in the same order as `label_descriptions`
    return label_descriptions[np.argmax(similarities)]
```

```python
df_embedding['assigned_label'] = df_embedding['embedding'].apply(lambda x: find_most_similar_label(x, label_embeddings))
```

```python
def model_1(df):
    """Model 1 is to determine whether inventory data exists and their location in each article"""

    response_msgs = []
      
    for _, row in df.iterrows():
        title = row[df.columns[0]]
        context = row['Table_title']
        
        
        system_msg = """
        Determine whether the provided context includes life cycle inventory data (or input-output data) related to LCA investigations. 
        """

        user_msg = """
        Context: 
        Table 3 Inputs data of methanol production referred to 1 kg of methanol
        Answer: Yes.

        Context: 
        Table 4 Outputs data of methanol production referred to 1 kg of methanol
        Answer: Yes.
        
        Context: 
        Table 2. Life Cycle Energy Consumption Inventory Data of the CtEG Route for Producing 1 ton EG 
        Answer: Yes.

        Context: 
        Table S2. Direct factors of pollutions. 
        Answer: No.

        Context: 
        Table 1 Key system design parameters. 
        Answer: No.

        Context: 
        Table S7 The inventory data for Coal -CH 3OH technical route. 
        Answer: Yes.

        Context: 
        TABLE 1 Main inventory data (per kg product) for ethanol and ethylene production 
        Answer: Yes.
        
        Context:
        Table. S3  The midpoint environmental impact s of the four typical ethylene glycol
        Answer: No.
        
        Context: 
        Table 2 Inventory analysis for propane ammoxidation scenarios. 
        Answer: Yes.
        """

        attempts = 3
        while attempts > 0:
            question = """Determine whether the provided context includes life cycle inventory data (or input-output data) of LCA, 
        and answer with either "Yes" or "No" only.
        """
            try:
                response = openai.chat.completions.create(
                    messages=[
                        {
                            "role": "system", 
                            "content": system_msg
                        },
                        {
                            "role": "user",
                            "content": user_msg + context + question
                        }
                    ],
                    model='gpt-3.5-turbo')
                
                answers = response.choices[0].message.content
                break

            except Exception as e:
                attempts -= 1
                if attempts > 0:
                    print(f"Error: {str(e)}. Retrying in 60 seconds. {attempts} attempts remaining. (model 1)")
                    time.sleep(60)
                else:
                    print(f"Error: Failed to process paper {title}. Skipping. (model 1)")
                    answers = "No"
                    break

        response_msgs.append(answers)
    df['results'] = response_msgs
    return df
```
